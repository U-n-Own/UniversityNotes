# Simulating Integrate and Fire model

```python
from integrate_fire import *
```

```python
neuron = IntegrateAndFireNeuron()
I = np.where((np.arange(0, 100, neuron.dt) > 10) & (np.arange(0, 100, neuron.dt) < 90), 20, neuron.resting_potential)  # Current is 20 between time 10 and 90, and resting_potential otherwise
t, u, spikes = neuron.simulate(I=I, T=100)

# Create a new figure
fig, axs = plt.subplots(2)

# Plot the membrane potential
axs[0].plot(t, u, label='Membrane potential')
axs[0].set_ylabel('Membrane potential')
axs[0].legend()

# Plot the spikes
axs[1].plot(t, spikes, label='Spikes')
axs[1].set_xlabel('Time')
axs[1].set_ylabel('Spikes')
axs[1].legend()

# Show the plot
plt.show()
```

```python
# Create three neurons with different initial conditions
neuron1 = IntegrateAndFireNeuron()
neuron1.u = 30  # Initial condition for neuron1

neuron2 = IntegrateAndFireNeuron()
neuron2.u = -60  # Initial condition for neuron2

neuron3 = IntegrateAndFireNeuron()
neuron3.u = -200 # Initial condition for neuron3

# Simulate the neurons
I = np.where((np.arange(0, 100, neuron1.dt) > 10) & (np.arange(0, 100, neuron1.dt) < 90), 20, neuron1.resting_potential)  # Current is 20 between time 10 and 90, and resting_potential otherwise
t1, u1, spikes1 = neuron1.simulate(I=I, T=100)
t2, u2, spikes2 = neuron2.simulate(I=I, T=100)
t3, u3, spikes3 = neuron3.simulate(I=I, T=100)

# Create a new figure
fig, axs = plt.subplots(2)

# Plot the membrane potentials
axs[0].plot(t1, u1, label='(neuron1)')
axs[0].plot(t2, u2, label='(neuron2)')
axs[0].plot(t3, u3, label='(neuron3)')
axs[0].set_ylabel('Membrane potential')
axs[0].legend()

# Plot the spikes
axs[1].plot(t1, spikes1, label='(neuron1)')
axs[1].plot(t2, spikes2, label='(neuron2)')
axs[1].plot(t3, spikes3, label='(neuron3)')
axs[1].set_xlabel('Time')
axs[1].set_ylabel('Spikes')
axs[1].legend()

# Show the plot
plt.show()
```
